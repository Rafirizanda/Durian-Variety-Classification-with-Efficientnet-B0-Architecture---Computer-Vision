```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from google.colab import drive, files

# 1. Mount Google Drive
drive.mount('/content/drive')
```

```python
# 2. Load model
model_path = "/content/drive/MyDrive/durian_split_7data/Model/model_terbaik_durian.h5"  # ubah sesuai path model
model = tf.keras.models.load_model(model_path)
```

```python
# Daftar kelas
class_names = ["bengkulu-durian", "d24(sultan)-durian", "goldenphoenix-durian", "kota_agung-durian", "medan-durian", "musang_king-durian", "sumatra_super-durian"]
```

```python
def predict_image(img_path, threshold=0.7):  # threshold bisa diatur, misal 0.7
    img = image.load_img(img_path, target_size=(224, 224, 3))  # sesuaikan ukuran input model
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)

    # Prediksi
    pred = model.predict(img_array)
    probabilities = pred[0]

    # Ambil kelas dengan probabilitas tertinggi
    class_index = np.argmax(probabilities)
    confidence = np.max(probabilities)

    # Tampilkan gambar
    plt.imshow(img)
    plt.axis("off")
    plt.show()

    # Cek apakah confidence cukup tinggi
    if confidence < threshold:
        print("Prediksi: Bukan Buah Durian yang dikenal (confidence rendah)")
    else:
        print(f"Prediksi: {class_names[class_index]} (Prob: {confidence:.2f})")

    # Print probabilitas tiap kelas
    print("\nProbabilitas tiap kelas:")
    for i, prob in enumerate(probabilities):
        print(f"- {class_names[i]}: {prob:.2f}")
```

```python
# 3. Upload gambar dan prediksi otomatis
# durian musang king
uploaded = files.upload()
for img_path in uploaded.keys():
    predict_image(img_path)
```
